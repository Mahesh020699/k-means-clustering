```python
# Import necessary libraries
import numpy as np
from collections import Counter
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
```

```python
#Loading and pre processing the Cifar10 Dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Defining the pre-trained model (VGG16)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Freeze some layers to prevent re-training
for layer in base_model.layers[-1:-12]:
    layer.trainable = False

model = models.Sequential()
model.add(base_model)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(10, activation='softmax'))

model.compile(optimizer=Adam(learning_rate=0.00005),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

#Using early stopping to stop being overfitted
es = EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True)

#Training the model using transfer learning
model.fit(x_train, y_train,
          epochs=50,
          validation_data=(x_test, y_test),
          callbacks=[es])

test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'Test accuracy: {test_acc}')
```

```python
#For finding out the layers which are frozen and unfrozen
frozen_layers = [layer.name for layer in model.layers if not layer.trainable]
unfrozen_layers = [layer.name for layer in model.layers if layer.trainable]

print("Frozen Layers:")
print(frozen_layers)

print("\nUnfrozen Layers:")
print(unfrozen_layers)
```

```python
# Building another model from scratch
model2 = models.Sequential()
model2.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model2.add(layers.MaxPooling2D((2, 2)))
model2.add(layers.Conv2D(64, (3, 3), activation='relu'))
model2.add(layers.MaxPooling2D((2, 2)))
model2.add(layers.Conv2D(64, (3, 3), activation='relu'))
model2.add(layers.Flatten())
model2.add(layers.Dense(64, activation='relu'))
model2.add(layers.Dense(10, activation='softmax'))

# Compiling the model
model2.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Training the model
model2.fit(x_train, y_train, epochs=15, validation_data=(x_test, y_test))

test_loss, test_acc = model2.evaluate(x_test, y_test)
print(f'Test accuracy: {test_acc}')
```

```python
#Defining the class names for better understanding
cifar10_class_names = [
    'airplane', 'automobile', 'bird', 'cat', 'deer',
    'dog', 'frog', 'horse', 'ship', 'truck'
]

label_to_index = {label: index for index, label in enumerate(cifar10_class_names)}

#Finding out the number of images in each class
train_labels = np.ravel(y_train)
label_counts = Counter(train_labels)
freq_series = pd.Series(label_counts)

print(freq_series)
x_train.shape
```

```python
# Displaying how the cifar-10 class looks
class_indices = {class_name: None for class_name in cifar10_class_names}

for class_name in cifar10_class_names:
    class_index = next(i for i, label in enumerate(y_train.flatten()) if label == cifar10_class_names.index(class_name))
    class_indices[class_name] = class_index

# Plotting one image for each class
fig, axes = plt.subplots(1, len(cifar10_class_names), figsize=(20, 2))
for i, class_name in enumerate(cifar10_class_names):
    axes[i].imshow(x_train[class_indices[class_name]])
    axes[i].set_title(class_name)
    axes[i].axis('off')

plt.show()
```

```python
# Evaluating and plotting predictions
def visualize_predictions(custom_model, images_to_test, labels_to_test, class_mapping, num_samples=25):

    # Predicting labels for the test set
    predicted_scores = custom_model.predict(images_to_test)
    predicted_labels = np.argmax(predicted_scores, axis=1)
    index_to_class = {index: class_name for class_name, index in class_mapping.items()}

    # Selecting random image indices from the test set
    random_indices = np.random.choice(len(images_to_test), size=num_samples, replace=False)

    # Displaying the images and predictions
    plt.figure(figsize=(11, 11))
    for idx, random_index in enumerate(random_indices):
        plt.subplot(5, 5, idx + 1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)

        # Denormalizing and converting to integers
        denormalized_image = (images_to_test[random_index] * 255).astype(int)

        # Plotting with improved image quality using interpolation
        plt.imshow(denormalized_image, interpolation='bilinear')
        predicted_label = index_to_class[predicted_labels[random_index]]
        true_label = index_to_class[labels_to_test[random_index][0]]

        color = 'green' if predicted_label == true_label else 'red'
        plt.xlabel(f"Predicted: {predicted_label}\nActual: {true_label}", color=color, fontsize=8)

    plt.show()

training_set, test_images, training_labels, test_labels = train_test_split(x_test, y_test, test_size=0.2, random_state=42)
```

```python
# Plot for model2 which is built from Scratch
visualize_predictions(model2, test_images, test_labels, label_to_index)
```

```python
# Plot for model which is built using transfer learning
visualize_predictions(model, test_images, test_labels, label_to_index)
```
